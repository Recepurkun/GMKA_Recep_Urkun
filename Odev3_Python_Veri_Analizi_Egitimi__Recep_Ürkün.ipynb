```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
from plotly.tools import make_subplots

import warnings
warnings.filterwarnings("ignore")
```

### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
#EUR/TRY Kur veri cekimi
Kur= yf.download("EURTRY=X", start= "2021-01-01", end = "2022-01-01")
Kur
```

2. Veriyi özet metodları ile inceleyiniz

```python
Kur.head(5)
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
Kur = Kur[['Adj Close']] #sadece istedigimiz sutunu gorebiliyoruz
Kur.columns = ["EUR/TRY"]
Kur
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
Kur['Normalize EUR/TRY'] = Kur["EUR/TRY"]/Kur.iloc[0,0] #ustteki yapilan islem ile tum degerleri norm'a gore boluyoruz boylece butun veriler pesi sira diziliyor.
Kur
```

```python
#BIST30 verilerini indirip duzenledik
B30 = yf.download("XU030.IS", start="2021-01-01", end="2021-12-31")
B30 = B30[["Adj Close"]]
B30.columns = ["Kapanis (TRY)"]
B30["Normalize"] = B30["Kapanis (TRY)"]/ B30.iloc[0,0]
B30
```

```python
#BIST30'un Normalize grafigi
fig_1 = px.line(B30, x=B30.index, y=B30.Normalize,
        width=1000, height=500,
        title = "BIST30 Endeksi 2021 - Normalize (1 ocak 2021)",
        hover_data = {"Normalize": ":.2f", "Kapanis (TRY)": ":.2f",},
        labels = {"Date" : "Tarih", "Normalize" : "BIST 30 Endeksi (TRY)"}
        )
fig_1.show()
```

# ***BIST 30 Endeksi 2021 yili sonlarına kadar standart inis cikis egilimleri gostermesine karsin yil sonuna dogru ivme kazanarak en yuksek 1.53 katina cikmistir. Seneyi ise 1.23 kat yukselerek kapatmistir.***


5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
#BIST30 ve Kur verilerini ayni tabloda gorebilmek icin birlestiriyoruz.
B30K_2021 = B30.merge(Kur, left_index= True, right_index=True)
B30K_2021
#Normalize EUR/TRY'nin 1den degil de 1.002309 olarak gozukme sebebi Borsanin 1.ayin 4unde acilmis olmasi.
```

```python
#B30k'nin px grafigi
px.line(B30K_2021, y=["Normalize","Normalize EUR/TRY"])
```

```python
#B30K'nin go. grafigi
fig = go.Figure()#oncelikle figur olusturuyoruz
fig.add_trace(go.Bar(x=B30K_2021.index, y= B30K_2021["Normalize"], name="BIST30")) #katman eklemek, bar grafigi cizdiriyoruz
fig.add_trace(go.Scatter(x=B30K_2021.index, y= B30K_2021["Normalize EUR/TRY"], name="EUR/TRY")) #cizgi grafigi cizdiriyoruz

fig.update_layout(title="Normalize BIST 100 ve Normalize Kur",
                  xaxis_title = "Tarih", xaxis_dtick="M1", #dtick ile butun aylari gostermesini sagliyoruz
                  yaxis_title = "Kapanis (TRY)")
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
indirme_listesi = ["XU100.IS","EURTRY=X"]
B30K_2022 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = "2022-01-01", end= "2023-01-01")
  data = data["Adj Close"].rename(i)
  B30K_2022 = pd.concat([B30K_2022, data], axis=1) #verilen 2 veriyi sutun olarak birlestirme islemi

B30K_2022
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
B30K_2022.columns = ["Kapanis (TRY)","EUR/TRY"]#ve satir isimlendirmelerini degistiriyoruz
B30K_2022 = B30K_2022[B30K_2022["Kapanis (TRY)"]>0]#borsanin kapali oldugu zaman dilimini tablomuzdan siliyoruz.
B30K_2022
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
B30K_2021 = B30K_2021[["Kapanis (TRY)","EUR/TRY"]]
B30K_2021
```

```python
Veri_21_22 = pd.concat([B30K_2021, B30K_2022], axis = 0)#2022 ile 2023 yillarinin verilerini birlestiriyoruz
Veri_21_22
```

```python
Veri_21_22["Normalize"] = Veri_21_22["Kapanis (TRY)"]/Veri_21_22.iloc[0,0]
Veri_21_22["Normalize EUR/TRY"] = Veri_21_22["EUR/TRY"]/Veri_21_22.loc[Veri_21_22.index[0],"EUR/TRY"]#once verinin indexini sec sonra da usd/try sutununu kullanmak istiyoruz diyoruz
Veri_21_22["degisim"] = Veri_21_22["Normalize"].diff(1)
Veri_21_22["d_yon"] = Veri_21_22["degisim"].map(lambda x: 'dusen gun' if x < 0 else 'yukselen gun') 
Veri_21_22["degisim%"] = (Veri_21_22["degisim"]/Veri_21_22["Normalize"].shift(1))*100
Veri_21_22
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
#PX GRAFIGI - B30 Normalize
fig_4 = px.line(Veri_21_22, y="Normalize",
        width = 1000, height=500,
        title="2021-2022 | BIST30 Gelişimi",
        hover_data={"Normalize" : ':.2f',"Kapanis (TRY)" : ":.0f"},
        labels = {"Normalize" : "Normalize (1 Ocak 2021 = 1)","index" : "Tarih"})
fig_4.show()
```

```python
#PX GRAFIGI - Normalize EUR/TRY
fig_5 = px.line(Veri_21_22, y="Normalize EUR/TRY",
        width = 1000, height=500,
        title="2021-2022 | EUR/TRY Gelişimi",
        hover_data={"EUR/TRY" : ':.2f',"Kapanis (TRY)" : ":.0f"},
        labels = {"Normalize" : "Normalize (1 Ocak 2021 = 1)","index" : "Tarih"})
fig_5.show()
```

```python
#px 2021-2022 | BIST30 ve EUR/TRY Gelişimi
px.line(Veri_21_22, y=["Normalize","Normalize EUR/TRY"])
```

```python
#GO. GRAFIGI
fig = go.Figure()
fig.add_trace(go.Bar(x=Veri_21_22.index, y= Veri_21_22["Normalize"], name="BIST30")) #katman eklemek, bar grafigi cizdiriyoruz
fig.add_trace(go.Scatter(x=Veri_21_22.index, y= Veri_21_22["Normalize EUR/TRY"], name="EUR/TRY")) #cizgi grafigi cizdiriyoruz

fig.update_layout(title="2021-20211 BIST30 ve EUR/TRY Gelişimi",
                  xaxis_title = "Tarih", xaxis_dtick="M1",
                  yaxis_title = "Normalize (1 Ocak 2021 = 1)")
```

```python
#subplots GRAFIGI
fig = make_subplots(rows=1, cols=2)
fig.add_trace(go.Scatter(x = Veri_21_22.index , y= Veri_21_22["Normalize"], name="Normalize"),
              row=1, col=1)
fig.add_trace(go.Bar(x = Veri_21_22.index , y= Veri_21_22["Normalize EUR/TRY"], name="Normalize EUR/TRY"),
              row=1, col=2)
fig.show()
```
